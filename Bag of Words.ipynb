# Bag of Words

Bag of words (bow) model is a way to preprocess text data for building machine learning models. In technical terms, we can say that it is a method of <b>feature extraction</b> with text data.

Natural language processing (NLP) uses bow technique to convert text documents to a machine understandable form. Each sentence is a document and words in the sentence are tokens. Count vectorizer creates a matrix with documents and token counts (bag of terms/tokens) therefore it is also known as document term matrix (dtm).

source: https://thatascience.com/learn-machine-learning/bag-of-words/#:~:text=Count%20vectorizer%20creates%20a%20matrix,document%20term%20matrix%20(dtm)

```python
# Imports 
import numpy as np
import pandas as pd
import os
from sklearn.feature_extraction.text import CountVectorizer
```

In this example, we use CountVectorizer() from the scikit-learn library to create bag of words

<i>CountVectorizer(): Converts a collection of text documents to a matrix of token counts</i>

source: https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.CountVectorizer.html

```python
# Let's create sample data
data = ['We are good',
        'We are becoming bemltter',
        'We will be great']
```

```python
# Instantiate count vectorizer
countvec = CountVectorizer()
cdf = countvec.fit_transform(data)
bow = pd.DataFrame(cdf.toarray(), columns = countvec.get_feature_names())
bow
```

options for feature extraction:
- bag of words
- TF-IDF (https://scikit-learn.org/stable/modules/generated/sklearn.feature_extraction.text.TfidfVectorizer.html)
